## Шаг 2. Предобработка данных

### Обработка пропусков и замена типа данных

````python
# для подстраховки от искажений перевожу в абсолютное значение значения столбца dob_years
data['dob_years'] = abs(data['dob_years'])

# шаблон задачи настаивает на замене пропущенных значений. ввожу переменную median_from_age, 
# рассчитывающую медианное значение трудового стажа в зависимости от возраста клиента.
# ставлю полученное из расчётов median_from_age в пустую ячейку столбца 'days_employed' 
# клиентам того же возраста
for age in data['dob_years'].unique():
    median = data.loc[data['dob_years'] == age, 'days_employed'].median()
    data.loc[(data['days_employed'].isna()) & (data['dob_years'] == age), \
    'days_employed'] = median

# аналогичную замену делаю для пропущенных значений в столбце total_income, отталкиваясь от 
# уровня образования клиента
for education_id in data['education_id'].unique():
    median_from_ed = data.loc[data['education_id'] == education_id, 'total_income'].median()
    data.loc[(data['total_income'].isna()) & (data['education_id'] == education_id), \
    'total_income'] = median_from_ed  
    
# перевожу тип даных из float в int
data['days_employed'] = data['days_employed'].astype('int')
data['total_income'] = data['total_income'].astype('int')    
    
# перевод значений в абсолютное значение, избавление от отрицательных значений
data['days_employed'] = abs(data['days_employed'])

# количество строк с неестественно большим/ошибочным значением дней трудового стажа, исходя из 
# максимума в 25000 дней
print('Строк со слишком большим трудовым стажем:',\
len(data.loc[data.loc[:,'days_employed'] > 25000])) 

# сведение всех символовв данных столбца education и family_status к нижнему регистру
data['education'] = data['education'].str.lower()
data['family_status'] = data['family_status'].str.lower()

# контроль выполнения кода
display(data.head(10))
data.info()
````

````python
# присвоение целям кредита сокращённой записи на основе самых популярных лемм 
# из прежних формулировок целей кредита
      
from pymystem3 import Mystem
m = Mystem() 

def purpose_func(purpose_lemmas):
    purposes = ['жилье', 'недвижимость', 'автомобиль', 'образование', 'свадьба']
    purpose_lemmas = m.lemmatize(purpose_lemmas['purpose'])
    for purpose in purposes:
        if purpose in purpose_lemmas:  
            return purpose
data['purpose'] = data.apply(purpose_func, axis=1)

# из-за практической идентичности понятий 'жилье' и 'недвижимость' переименовываю цели, 
# таким образом объединяя эти одинаковые цели в одну
data.loc[data['purpose'] == 'жилье', 'purpose'] = 'недвижимость' 

# подсчёт количества целей=лемм, контроль выполнения последних действий по объединению целей
from collections import Counter
print(Counter(data.apply(purpose_func, axis=1))) 
````

Шаблон решения проектной работы обязывает использование лемматизации, что, по моему мнению, в этой задаче не нужно, т.к. формулировок целей немного - 38, разбивка в ручную, и присвоение в цикле в этом случае даст разбивку на подцели гораздо точнее.
